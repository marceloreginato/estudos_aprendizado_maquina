# Carregando e preparando dados para o exercício

No exercício desta semana, o objetivo é comparar três algoritmos de classificação utilizando um conjunto de dados de [digitos escritos a mão](https://archive.ics.uci.edu/dataset/80/optical+recognition+of+handwritten+digits). Neste conjunto de dados, cada digito é representado por uma matriz (imagem) 8x8. No entanto, iremos tratar cada posição da matriz como um atributo ao invés de classificar a imagem como um todo.  

A seguir, os dados serão carregados e processados de forma a criar conjuntos de treino e teste.

```python
# Importando bibliotecas
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier, BaggingClassifier
from sklearn.neural_network import MLPClassifier
```

```python
# Carregando os dados
raw = load_digits(as_frame=True)
```

```python
# print(raw.DESCR)
```

```python
# Separando features e target
X = raw.data.values
y = raw.target
```

```python
# raw.data.info()
```

```python
# Separando os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2026, stratify=y)
```


# 1) Árvores de Decisão

Avalie [árvores de decisão](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html) utilizando os conjuntos de treino e teste criados anteriormente e  configurando a *seed* para o valor 2026. Reporte **acurácia**, **precisão** e **revocação (sensibilidade)** para os dados de teste. Para este exercício, **avalie três configurações** de parâmetros para o modelo. Configure a *seed* para o valor 2026.

```python
# Função para modularizar as Árvores
def configDecisionTree(max_depth=None):
  # Instanciando modelo
  dt = DecisionTreeClassifier(max_depth=max_depth, random_state=2026)
  # Treinando e realizando previsões
  dt.fit(X_train, y_train)
  y_pred = dt.predict(X_test)
  return y_pred

def avaliaModelo(y_pred, y_test, max_depth=None):
  # Cálculo das métricas
  accuracy = accuracy_score(y_test, y_pred)
  recall = recall_score(y_test, y_pred, average = 'macro')
  precision = precision_score(y_test, y_pred, average = 'macro')
  # Printando métricas
  print(f'Configuração {max_depth}')
  print('Acurácia:', accuracy)
  print('Revocação:', recall)
  print('Precisão:', precision)
  print()

y_pred = configDecisionTree(10)
avaliaModelo(y_pred, y_test, 10)

y_pred = configDecisionTree(30)
avaliaModelo(y_pred, y_test, 30)

y_pred = configDecisionTree(50)
avaliaModelo(y_pred, y_test, 50)
```

1. Um método de ensemble em aprendizado de máquina consiste na combinação de múltiplos modelos para gerar uma predição final, com o objetivo de obter desempenho superior ao de um único modelo. A ideia é reduzir erros, aumentar a robustez e melhorar a capacidade de generalização do sistema. A Random Forest corresponde a um método de ensemble do tipo bagging (Bootstrap Aggregating), no qual várias árvores de decisão são treinadas em subconjuntos aleatórios dos dados gerados por amostragem com reposição. A predição final é obtida pela agregação das previsões individuais.

2. O bootstrapping afeta o funcionamento das árvores aleatórias ao fazer com que cada árvore da Random Forest seja treinada em um subconjunto diferente dos dados, obtido por amostragem aleatória com reposição. Isso significa que alguns exemplos podem aparecer repetidos em um conjunto de treino, enquanto outros podem nem ser selecionados.
Esse processo aumenta a diversidade entre as árvores, fazendo com que cada uma aprenda padrões ligeiramente diferentes. Como resultado, a combinação das previsões reduz a variância do modelo e diminui o risco de overfitting, tornando a Random Forest mais robusta do que uma única árvore de decisão.

3. A principal diferença entre bagging e boosting está na forma como os modelos são treinados e combinados. No bagging, os modelos são treinados de forma independente e paralela, utilizando diferentes subconjuntos aleatórios dos dados. O objetivo é reduzir a variância do modelo, sendo a Random Forest um exemplo desse método. Já no boosting, os modelos são treinados de forma sequencial, onde cada novo modelo busca corrigir os erros cometidos pelos anteriores. Nesse caso, exemplos difíceis de classificar recebem maior atenção ao longo do treinamento. O objetivo é reduzir o viés e melhorar gradualmente o desempenho, como ocorre em métodos como AdaBoost e Gradient Boosting.

4. A ideia central do gradient boosting é construir modelos de forma sequencial, em que cada novo modelo é treinado para corrigir os erros do modelo anterior, minimizando gradualmente uma função de perda por meio do gradiente. Em geral, árvores fracas são adicionadas sucessivamente para melhorar a predição do conjunto. Isso difere do método usado pelas Random Forests, que utilizam bagging. Nas florestas aleatórias, as árvores são treinadas independentemente e em paralelo, sobre subconjuntos aleatórios dos dados, e a predição final é feita por agregação (votação ou média).

5. No contexto de gradient boosting, o gradiente representa a direção de maior redução da função de perda, indicando os erros que o modelo atual ainda está cometendo. A cada etapa, um novo modelo é treinado para aprender esses erros, isto é, aproximar os resíduos gerados pelo modelo anterior.
