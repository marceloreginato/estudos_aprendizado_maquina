## Carregando dados

Para este exercício, será utilizado o conjunto de dados *Breast Cancer* disponibilizado pelo [sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_breast_cancer.html).

```python
# Importando bibliotecas
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import accuracy_score, recall_score
from sklearn.linear_model import Perceptron
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
raw_data = load_breast_cancer()
```

```python
df = pd.DataFrame(raw_data.data, columns=raw_data.feature_names)
labels = raw_data.target
```

## 1) Perceptron (regressão logística) e avaliação de performance

Neste exercício, o objetivo é investigar o comportamento do modelo Perceptron, suas limitações e avaliá-lo em um exemplo de problema de classificação. Além disto, serão abordadas duas estratégias diferentes de avaliação: *holdout* e *K-Folds*.

- a) Divida o os dados em treino e teste usando a função [*train_test_split*](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) do sklearn. Esta função implementa a avaliação baseada em *holdout*. Utilize 20% dos dados para avaliação, e a seed 2026 para a divisão dos dados. Estratifique os conjuntos com base nos rótulos (veja na documentação o significado do parâmetro '*stratify*').
- b) Padronize os dados como pré-processamento.
- c) Compute a acurácia no conjunto de teste e a medida revocação, utilizando o [Perceptron](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Perceptron.html) com seus parâmetros padrão. Apenas configure sua *seed* para a estabelecida na divisão de treino e teste.
- d) Repita os itens anteriores, utilizando uma transformação polinomial das *features* originais. Utilize o [PolynomialFeatures](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html#sklearn.preprocessing.PolynomialFeatures) (usando "*degree*" igual a 2) como pré-processamento no lugar do StandardScaler.
- e) Repita a avaliação usando uma validação cruzada de 5-folds estratificada (use [K-folds](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedKFold.html) ou [*cross_val_score*](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html)).  **Para simplificar o exercício, utilize os dados sem pré-processamento.**


```python
# a) DIVIDINO EM TREINO E TESTE
# Atribuindo as featues à X e o target à y
X = df
y = labels
# Função para separar os dados
# O stratify tenta utilizar uma proporção para separar os dados de treino e teste, de maneira que se a classe dominante é de 80% e a minoritária seja 20%, o stratify tenta manter essa mesma
# proporção tanto no conjunto de treino como no de teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state = 2026, stratify = y)

# b) PADRONIZANDO OS DADOS
# Instanciando o scaler para padronizar os dados
scaler_standard = StandardScaler()
# Aprende a escala com os dados de treino
X_train_scaled = scaler_standard.fit_transform(X_train)
# Aplica a mesma escala dos dados de treino nos dados de teste
X_test_scaled = scaler_standard.transform(X_test)

# c) PERCEPTRON PARÂMETROS PADRÃO
percep_std = Perceptron(random_state = 2026)
# Treinando modelo
percep_std.fit(X_train_scaled, y_train)
# Prevendo os valores
y_pred = percep_std.predict(X_test_scaled)
# Calculando métricas (QUERO ENTENDER MELHOR COMO FUNCIONA OS PARÂMETROS DO RECALL)
acc_percep = accuracy_score(y_test, y_pred)
rec_percep = recall_score(y_test, y_pred, pos_label = 0)
# Printando métricas
print("Acurácia para pré-processamento padrão:", acc_percep)
print("Recall para pré-processamento padrão:", rec_percep)

# d) ITEM b) e c) ANTERIORES PADRONIZANDO COM POLYNOMIAL FEATURES
scaler_polynomial = PolynomialFeatures(degree = 2)
# Pré-processamento nos dados de treino, aprendendo escala dos dados de treino
X_train_scaled = scaler_polynomial.fit_transform(X_train)
# Aplicando a mesma escala dos dados de treino nos dados de teste
X_test_scaled = scaler_polynomial.transform(X_test)
# Aplicando no Modelo
percep_poly = Perceptron(random_state = 2026)
# Treinando o modelo
percep_poly.fit(X_train_scaled, y_train)
# Prevendo
y_pred = percep_poly.predict(X_test_scaled)
# Calculando métricas (QUERO ENTENDER MELHOR COMO FUNCIONA OS PARÂMETROS DO RECALL)
acc_percep = accuracy_score(y_test, y_pred)
rec_percep = recall_score(y_test, y_pred, pos_label = 0)
# Printando métricas
print("\n")
print("Acurácia para pré-processamento polinomial:", acc_percep)
print("Recall para pré-processamento polinomial:", rec_percep)

# e) UTILIZAÇÃO DO CROSS_VAL_SCORE
percep_cvs = Perceptron(random_state=2026)
# Comando cross_val_score já separa em grupos treina, preve e calcula métrica
rec_cvs = cross_val_score(percep_cvs, X, y, cv=5, scoring="recall")
acc_cvs = cross_val_score(percep_cvs, X, y, cv=5, scoring="accuracy")
# Printando métricas recall
print("\nUtilizando Cross_val_score para recall:")
for i, recall in enumerate(rec_cvs):
  print(f"Recall grupo {i}: {recall}")
# Printando métricas acurácia
print("\nUtilizando Cross_val_score para acurácia:")
for i, acc in enumerate(acc_cvs):
  print(f"Acurácia grupo {i}: {acc}")
```

## 2) Multi-Layer Perceptron

Dando continuidade, este exercício foca na estudo do modelo MLP.

- a) Repita os 3 primeiros itens do exercício anterior, avaliando uma MLP no lugar do Perceptron. Utilize os parâmetros padrão da [MLP](https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html) do sklearn inicialmente e a *seed* igual a 2026.
- b) Avalie a MLP, usando a validação *holdout* dos exercícios anteriores, utilizando 3 configurações de parâmetos diferentes. Dê preferência para modificar número de neurônios e camadas.
- c) Compare a melhor MLP deste exercício, com o melhor Perceptron do anterior utilizando uma validação cruzada estratificada de 5-folds. **Por simplicidade, utilize dados sem pré-processamento.**


```python
# a) RESOLVENDO OS 3 PRIMEIROS ITENS ANTERIORES, PORÉM COM MLP
# Aprende a escala com os dados de treino
X_train_scaled = scaler_standard.fit_transform(X_train)
# Aplica a mesma escala dos dados de treino nos dados de teste
X_test_scaled = scaler_standard.transform(X_test)
# Instanciando Modelo
MLP_std = MLPClassifier(random_state=2026)
# Treinando e prevendo
MLP_std.fit(X_train_scaled, y_train)
y_pred = MLP_std.predict(X_test_scaled)
# Calculando métricas
acc_mlp = accuracy_score(y_test, y_pred)
rec_mlp = recall_score(y_test, y_pred, pos_label = 0)
# Printando métricas
print("Acurácia para MLP pré-processamento padrão:", acc_mlp)
print("Recall para MLP pré-processamento padrão:", rec_mlp)


# b) AVALIANDO PARA MLPs COM ARQUITETURAS DIFERENTES
# Instanciando Modelos
MLP_1 = MLPClassifier(hidden_layer_sizes=(10,), random_state=2026)
MLP_2 = MLPClassifier(hidden_layer_sizes=(10,10), random_state=2026)
MLP_3 = MLPClassifier(hidden_layer_sizes=(5,10,5), random_state=2026)
# Treinando e prevendo
MLP_1.fit(X_train_scaled, y_train)
MLP_2.fit(X_train_scaled, y_train)
MLP_3.fit(X_train_scaled, y_train)
y_pred_1 = MLP_1.predict(X_test_scaled)
y_pred_2 = MLP_2.predict(X_test_scaled)
y_pred_3 = MLP_3.predict(X_test_scaled)
# Calculando métricas
acc_mlp_1 = accuracy_score(y_test, y_pred_1)
rec_mlp_1 = recall_score(y_test, y_pred_1, pos_label = 0)
acc_mlp_2 = accuracy_score(y_test, y_pred_2)
rec_mlp_2 = recall_score(y_test, y_pred_2, pos_label = 0)
acc_mlp_3 = accuracy_score(y_test, y_pred_3)
rec_mlp_3 = recall_score(y_test, y_pred_3, pos_label = 0)
# Printando métricas
print("Acurácia para MLP_1 pré-processamento padrão:", acc_mlp_1)
print("Recall para MLP_1 pré-processamento padrão:", rec_mlp_1)
print("Acurácia para MLP_2 pré-processamento padrão:", acc_mlp_2)
print("Recall para MLP_2 pré-processamento padrão:", rec_mlp_2)
print("Acurácia para MLP_3 pré-processamento padrão:", acc_mlp_3)
print("Recall para MLP_3 pré-processamento padrão:", rec_mlp_3)

# c) Não entendi muito bem o que é pra fazer
# Acredito que em média comparando com o perceptron do exercício anterior a MLP_2 obteve melhores métricas de avaliação
```

## 3) Seleção de parâmetros

Note que no exercício anterior, foram testados diferentes configurações de MLP de forma manual. Neste exercício, o foco será em testar diferentes configurações do modelo de forma automática e reportar a melhor configuração.

- a) Leia a documentação do [GridSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html).
- b) Configure o "*param_grid*" com ao menos dois parâmetros (Ex: '*hidden_layer_sizes*' e '*activation*').
- c) Escolha uma métrica de classificação para comparar as MLPs e execute o GridSearchCV. **Para simplificar o exercício, utilize os dados sem pré-processamento no 'fit'.**
- d) Indique a melhor configuração obtida em termos da métrica escolhida.

```python
# ENTENDER MELHOR DEPOIS
MLP = MLPClassifier(random_state=2026)
parametros_malha = {'hidden_layer_sizes':[(10,), (10,10), (5,10,5)],
                     'activation': ['relu', 'tanh']}
grid_search = GridSearchCV(estimator=MLP, param_grid=parametros_malha, scoring='recall')
grid_search.fit(X_train, y_train)
grid_search.best_estimator_
y_pred = grid_search.predict(X_test)
rec = recall_score(y_test, y_pred, pos_label = 0)
print(rec)
```
