## 1. Pré-processamento

Realize o pré-processamento dos dados acima considerando os seguintes etapas.


1.   Valide que não existem valores nulos dentre os dados carregados.
2.   Separe os atributos da base de dados em **numéricos** e **categóricos**.
3.   Descarte os atributos categóricos.
4.   Isole a variável a ser predita "price".
4.   Separe os dados em treino e teste, considerando 20% dos dados para teste e usando a seed 2026.
5.   Finalize padronizando os dados divididos.



```python
# 1. VALIDAÇÃO DE VALORES NULOS
print(df.isnull().sum())

# 2. SEPARAÇÃO DOS ATRIBUTOS EM CATEGÓRICOS E NUMÉRICOS
df_num = df.select_dtypes(include=['int64', 'float64'])
df_cat = df.select_dtypes(include=['object'])

# 3. DESCARTE ATRIBUTOS CATEGÓRICOS
df = df.drop(columns = df_cat.columns)

# 4. ISOLAR TARGET
y = df['price']
X = df.drop(columns = ['price'])

# 5. SEPARANDO OS DADOS EM TREINO E TESTE -> (NÃO É POSSÍVEL USAR STRATIFY POIS É UM PROBLEMA DE REGRESSÃO)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2026)

# 6. PADRONIZANDO DADOS
scaler = StandardScaler()
# Aprende a escala com os dados de treino
X_train_scaled = scaler.fit_transform(X_train)
# Aplica a mesma escala dos dados de treino nos dados de teste
X_test_scaled = scaler.transform(X_test)
```

## 2. Avaliação de uma abordagem "ingênua"

 Leia a documentação do [*DummyRegressor*](https://scikit-learn.org/stable/modules/generated/sklearn.dummy.DummyRegressor.html) e o treine nos dados processados usando seus parâmetros padrões. Avalie a performance do modelo usando a métrica MSE.

```python
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Dummy regressor com seus parâmetros padrões
dummy_reg = DummyRegressor()
# Treinando e prevendo
dummy_reg.fit(X_train_scaled, y_train)
y_pred = dummy_reg.predict(X_test_scaled)
# Cálculo da métrica MSE
mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)
```

## 3. Comparação com outros métodos de regressão.

Agora iremos comparar o modelo "ingênuo" do exercício anterior com outros métodos de regressão.

1.  Escolha, treine e avalie dois métodos de regressão implementados pelo sklearn. Use a documentação do [sklearn](https://scikit-learn.org/stable/supervised_learning.html) para escolher os modelos e também utilize a mesma métrica e os mesmos dados de treino e teste do exercício anterior.
2.  Compare os resultados com o modelo anterior e justifique o desempenho do **pior modelo**.

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
# Regressor linear com seus parâmetros padrões
lin_reg = LinearRegression()
# Treinando e prevendo
lin_reg.fit(X_train_scaled, y_train)
y_pred = lin_reg.predict(X_test_scaled)
# Calculando métrica
mse = mean_squared_error(y_test, y_pred)
print("MSE Linear regression:", mse)

# Arvore de decisão regressora com seus parâmetros padrões
decision_tree_reg = DecisionTreeRegressor(max_depth= 3)
decision_tree_reg.fit(X_train_scaled, y_train)
# Treinando e predizendo
y_pred = decision_tree_reg.predict(X_test_scaled)
# Cálculo métrica
mse = mean_squared_error(y_test, y_pred)
print("MSE Decision Tree Regression:", mse)

# O pior modelo foi o dummy regressor pois como ele apenas calcula a média, obteve erros maiores
# ao predizer os valores de y, o fato da regressão linear ter tido um erro menor que a árvore de decisão
# aponta que deve haver uma relação linear entre o target e as features.
```
